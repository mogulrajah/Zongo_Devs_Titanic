```python
#dropping the following columns name,ticket and cabin
train.drop(columns=['Name', 'Ticket', 'Cabin','PassengerId'],inplace=True)
train
```

```python
#finding the mode of the embarked column
Mode=train['Embarked'].mode()
```

```python
Mode
```

```python
#using fillna to fill the empty enrties in Embarked
train['Embarked'].fillna(value='S',inplace=True)
```

```python
#checking if our fillna function filled the enpty entries in our embarked column
train['Embarked'].isna().sum()
```

```python
train
```

```python
#Finding the median age using the passenger class
train.groupby('Pclass')['Age'].median()
```

```python
#Using the median age of the various passenger class to fill our age columns
train.loc[(train['Pclass'] == 1) & (train['Age'].isna()),'Age'] =37

train.loc[(train['Pclass'] == 2) & (train['Age'].isna()),'Age'] =29

train.loc[(train['Pclass'] == 3) & (train['Age'].isna()),'Age'] =24
```

```python
# checking the First 50 entries in our dataset after finding the median age and filling our data
train.head(50)
```

```python
#Using the isna.sum to check if there are empty entries in our dataset after filling
train.isna().sum()
```

```python
#Encoding the sex column
encode = LabelEncoder()
train['Sex']=encode.fit_transform(train["Sex"])
train
```

```python
#Encoding the embarked column
encode = LabelEncoder()
train['Embarked']=encode.fit_transform(train["Embarked"])
train
```

```python
# Count the occurrences of 0 and 1 in the 'Survived' column
survival_counts = train['Survived'].value_counts()

print(survival_counts)
```

```python
# Process the data to get the counts of survivors by gender
gsurvival = train[train['Survived'] == 1]['Sex'].value_counts()
# Create a pie chart
plt.pie(gsurvival, labels=gsurvival.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("Survivors by Gender")
plt.show()
```

```python
#doing a feature labeling on our dataset
from sklearn.preprocessing import StandardScaler
# initializing the scaler and scaling our age and sex column
scaler = StandardScaler()
train[['Age','Sex']] = scaler.fit_transform(train[['Age','Sex']])
# output of sex and age after feature labeliing
train
```

```python
# splitting our dateset into test and train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train[['Age','Sex']],train['Survived'],test_size=0.2)
```
